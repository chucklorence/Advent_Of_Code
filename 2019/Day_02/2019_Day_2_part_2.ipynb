```python
import numpy as np

###########################################################################################
# Global Variables
###########################################################################################
# Global computer variable
intcode_program = np.array([])
# Keeps track of the number of times we've worked an iteration of the program
iteration_counter = 0
# Indicates our location in the program
position_pointer = 0
# Log file name
log_file_name = "log_file.txt"


###########################################################################################
# Functions
###########################################################################################
# Get the original puzzel inputs
def Get_Computer_Inputs():    
    file_contents = open("puzzel_input.txt","r").read()    
    inputs = [int(num.strip()) for num in file_contents.split(",") ]
    return inputs

# Save state
def Save_State():
    return

# Returns the next 4 values from the intcode program
def Get_Next_4():
    next4 = intcode_program[position_pointer:position_pointer+4]
    return next4
    
#
def Increment_Pointer():
    global position_pointer
    position_pointer = position_pointer + 4
    #print(f"position pointer = {position_pointer}")
    return

#
def Reset_Pointer():
    global position_pointer
    position_pointer = 0
    #print(f"position pointer = {position_pointer}")
    return

#
def Increment_Counter():
    global iteration_counter
    iteration_counter = iteration_counter + 1
    #print(f"iteration_counter = {iteration_counter}")
    return

# Log iteration for debugging purposes 
def Log_Iteration(program_segment):
    file = open(log_file_name,"a")    
    if iteration_counter == 1:
        file.write(f"({iteration_counter}; Segment = [{','.join([str(num) for num in program_segment])}];  pointer: {position_pointer}\n")
    else:
        file.write(f"\n({iteration_counter}; Segment = [{','.join([str(num) for num in program_segment])}];  pointer: {position_pointer}\n")
    
    file.close()
    with open(log_file_name,"a") as file:        
        file.write(','.join([str(num) for num in intcode_program]))
        file.write('\n')
    return

#
def Clear_Log():
    file = open(log_file_name,"w")
    file.write("")
    file.close()
    return   
```

```python
############################################################################
# Determine what pair of inputs produces the output 19690720
# The answer will be noun=X and verb=Y --> nounVerb
# E.g. noun=12 and very=2 = 1202
############################################################################

# Create the Lists of lists where each list contains a pair of numbers we want to use to find the answer
input_pairs = []
for noun in range(0,100):
    for verb in range(0,100):
        input_pairs.append([noun,verb])

# Reset our log file
Clear_Log()

# Store the final pair of numbers
final_input_pair = []

for input_pair in input_pairs:
    
    # Reset the pointer
    Reset_Pointer()

    # Refresh the intcode_program
    intcode_program = Get_Computer_Inputs()

    # Replace values at addresses 1 and 2
    intcode_program[1] = input_pair[0]
    intcode_program[2] = input_pair[1]    

    continue_program = True
    while continue_program == True:

        Increment_Counter()
        next4 = Get_Next_4()

        if next4[0] == 1:
            #print('Addition')   
            loc1 = next4[1]
            loc2 = next4[2]
            locStore = next4[3]
            result = intcode_program[loc1] + intcode_program[loc2]
            intcode_program[locStore] = result
            continue_program = True
        elif next4[0] == 2:
            #print('Multiplication')
            loc1 = next4[1]
            loc2 = next4[2]
            locStore = next4[3]
            result = intcode_program[loc1] * intcode_program[loc2]
            intcode_program[locStore] = result        
            continue_program = True
        elif next4[0] == 99:
            #print('End Program')
            continue_program = False
        else:
            print('An Error has occured')
            continue_program = False
        
        #Log_Iteration(next4)
        if continue_program == True:
            Increment_Pointer()
    
    # Program has completed
    # Check the result at position 0 to determine if we have the result we're looking for
    if intcode_program[0] == 19690720:
        final_input_pair = input_pair
        break
    #else:
    #    print(f"Program halted. Not the value we're looking for: {intcode_program[0]}")

# Check to see if we found the final pair
if len(final_input_pair) == 0:
    print("Pair NOT found.")
else:
    print(f"Program halted. Found it! Value at position 0: {intcode_program[0]}")
    print(f"Noun = {final_input_pair[0]}. Verb = {final_input_pair[1]}.")





```
